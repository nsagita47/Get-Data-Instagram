```
import instaloader
import emoji
import pandas as pd
import numpy as np
import datetime as dt
```

```
instagram = instaloader.Instaloader()
Scarlet = instaloader.Profile.from_username(instagram.context, 'scarlett_whitening')
```

```
for post in Scarlet.get_posts():
  print(post.caption)
```

```
day_name = ["Senin", "Selasa", "Rabu", "Kamis", "Jumat", "Sabtu", "Minggu"]
data = []
for post in Scarlet.get_posts():
  date = post.date_local
  if  date > dt.datetime(2020,8,30):
    caption = post.caption
    tipe = post.typename
    url = "www.instagram.com/p/"+post.shortcode
    likes = post.likes
    comments = post.comments
    engagements = (0.75*likes) + comments
    number_of_hastag = len([x for x in caption.split() if x.startswith('#')])
    number_of_emojis = len([x for x in caption.split() if x in emoji.UNICODE_EMOJI])
    hari = date.weekday()
    hari = day_name[hari]
    jam = date.hour
    date = date.strftime("%y-%m-%d %H:%M:%S")
    data.append([date, hari, jam, url, caption, tipe, number_of_hastag, number_of_emojis, likes,comments, engagements])
  else :
    pass
  data_ig = pd.DataFrame(data, columns =["date", "hari", "jam", "url", "caption", "tipe", "number_of_hastag", "number_of_emojis", "likes","comments", "engagements"])
```
